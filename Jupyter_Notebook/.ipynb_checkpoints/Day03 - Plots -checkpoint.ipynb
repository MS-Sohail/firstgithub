```python
# steps involved in Data Visualization

# Step-1 import libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step-2 set a theme

sns.set_theme(style="ticks", color_codes=True)
```

```python
# Step-3 import dataset, you can also import your own data

kashti = sns.load_dataset("titanic")
```

```python
# step-4 plot basic graph with online 1 variable

p = sns.countplot(x="sex", data=kashti)
plt.show()
```

```python
# step-5 plot basic graph with two variables

p = sns.countplot(x="sex", data=kashti, hue="class")
plt.show()
```

```python
# step-6 plot basic graph with two variables having Title

p = sns.countplot(x="sex", data=kashti, hue="class")
p.set_title("Visualization with Sohail")
plt.show()
```
